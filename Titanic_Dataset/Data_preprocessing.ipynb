# Performing Data Cleaning and Analysis

```python
titanic.describe() # Descriptive statistics
```

```python
# Name column can never decide survival of a person, hence we can safely delete it
del titanic["Name"] # delete the name column
titanic.head()
```

```python
del titanic["Ticket"] # delete the ticket column
titanic.head()
```

```python
del titanic["Fare"] # delete the fare column
titanic.head()
```

```python
del titanic['Cabin'] # delete the cabin column
titanic.head()
```

```python
# Changing Value for "Male, Female" string values to numeric values , male=1 and female=2
def getNumber(str):  # def function
    if str=="male":  # conditional statement
        return 1
    else:
        return 2
titanic["Gender"]=titanic["Sex"].apply(getNumber) #created a new column
#We have created a new column called "Gender" and
#filling it with values 1,2 based on the values of sex column
titanic.head()
```

```python
#Deleting Sex column, since no use of it now
del titanic["Sex"]
titanic.head()
```

```python
titanic.isnull().sum() # Check missing value
```

```python
meanS= titanic[titanic.Survived==1].Age.mean()
meanS
```

# Creating a new "Age" column , filling values in it with a condition if goes True then given values (here meanS) is put in place of last values else nothing happens, simply the values are copied from the "Age" column of the dataset

```python
titanic["age"]=np.where(pd.isnull(titanic.Age) & titanic["Survived"]==1  ,meanS, titanic["Age"])
titanic.head()
```

```python
titanic.isnull().sum() # see the missing values
```

```python
# Finding the mean age of "Not Survived" people
meanNS=titanic[titanic.Survived==0].Age.mean()
meanNS
```

```python
titanic.age.fillna(meanNS,inplace=True)
titanic.head()
```

```python
titanic.isnull().sum() # check missing value
```

```python
del titanic['Age'] # delete the Age column
titanic.head()
```

# We want to check if "Embarked" column is is important for analysis or not, that is whether survival of the person depends on the Embarked column value or not

```python
# Finding the number of people who have survived
# given that they have embarked or boarded from a particular port

survivedQ = titanic[titanic.Embarked == 'Q'][titanic.Survived == 1].shape[0]
survivedC = titanic[titanic.Embarked == 'C'][titanic.Survived == 1].shape[0]
survivedS = titanic[titanic.Embarked == 'S'][titanic.Survived == 1].shape[0]
print(survivedQ)
print(survivedC)
print(survivedS)
```

```python
survivedQ = titanic[titanic.Embarked == 'Q'][titanic.Survived == 0].shape[0]
survivedC = titanic[titanic.Embarked == 'C'][titanic.Survived == 0].shape[0]
survivedS = titanic[titanic.Embarked == 'S'][titanic.Survived == 0].shape[0]
print(survivedQ)
print(survivedC)
print(survivedS)
```

```python
titanic.dropna(inplace=True) #drop the missing value
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
#Renaming "age" and "gender" columns
titanic.rename(columns={'age':'Age'}, inplace=True)
titanic.head()
```

```python
titanic.rename(columns={'Gender':'Sex'}, inplace=True) # rename the gender column to sex
titanic.head()
```

```python
def getEmb(str):
    if str=="S":
        return 1
    elif str=='Q':
        return 2
    else:
        return 3
titanic["Embark"]=titanic["Embarked"].apply(getEmb)
titanic.head()
```

```python
del titanic['Embarked']
titanic.rename(columns={'Embark':'Embarked'}, inplace=True)
titanic.head()
```

# Visualization

```python
#Drawing a pie chart for number of males and females aboard
import matplotlib.pyplot as plt
from matplotlib import style

males = (titanic['Sex'] == 1).sum()
#Summing up all the values of column gender with a
#condition for male and similary for females
females = (titanic['Sex'] == 2).sum()
print(males)
print(females)
p = [males, females]
plt.pie(p,    #giving array
       labels = ['Male', 'Female'], #Correspndingly giving labels
       colors = ['green', 'yellow'],   # Corresponding colors
       explode = (0.15, 0),    #How much the gap should me there between the pies
       startangle = 0)  #what start angle should be given
plt.axis('equal')
plt.show()
```

```python
# More Precise Pie Chart
MaleS=titanic[titanic.Sex==1][titanic.Survived==1].shape[0]
print(MaleS)
MaleN=titanic[titanic.Sex==1][titanic.Survived==0].shape[0]
print(MaleN)
FemaleS=titanic[titanic.Sex==2][titanic.Survived==1].shape[0]
print(FemaleS)
FemaleN=titanic[titanic.Sex==2][titanic.Survived==0].shape[0]
print(FemaleN)
```

```python
chart=[MaleS,MaleN,FemaleS,FemaleN]
colors=['lightskyblue','yellowgreen','Yellow','Orange']
labels=["Survived Male","Not Survived Male","Survived Female","Not Survived Female"]
explode=[0,0.05,0,0.1]
plt.pie(chart,labels=labels,colors=colors,explode=explode,startangle=100,counterclock=False,autopct="%.2f%%")
plt.axis("equal")
plt.show()
```

**Insight:** The pie chart that displays the proportion of males and females on board gives a quick overview of the gender distribution of passengers.

**Further insight:** The more precise pie chart segmenting survival by gender provides a crucial insight – a significantly larger proportion of females survived compared to males. This suggests that gender might have played a role in survival.

Females survival is more than males

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(titanic['Age'], bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**Insight:** A histogram of passenger ages (as suggested in my previous response) would reveal the age distribution, possibly showing if there were more children, adults, or elderly passengers on board.

```python
sns.boxplot(x='Survived', y='Age', data=titanic)
plt.title('Age vs. Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

**Insight:** A box plot comparing the age distribution of survivors and non-survivors (also suggested earlier) would highlight any differences in age between the two groups.

**Further insight:** For instance, it might reveal if children had a higher survival rate or if a particular age range had lower chances of survival.

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

**Insight:** Count plots for survival by passenger class or embarked port show the number of survivors and non-survivors for each category.

**Further insight:** They can reveal if passengers in certain classes (e.g., first class) or embarked from specific ports had better survival odds. This could indicate factors like access to lifeboats or socioeconomic differences influencing survival.
